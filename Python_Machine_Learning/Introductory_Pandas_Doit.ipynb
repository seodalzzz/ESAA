###2) 다변량 그래프 그리기
  * 다변량 그래프 : 3개 이상의 변수를 사용한 그래프
  * 다변량 그래프가 필요한 이유
    * 3개 이상의 변수를 사용하는 다변량 그래프는 적재적소에 맞는 요소를 추가하여 그래프를 표현해야 함

###<다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프>


####(1) 앞의 산점도 그래프에 성별을 새 변수로 추가
  * 문자열은 산점도 그래프의 색상을 지정하는 값으로 사용할 수 없음 >> 0,1과 같은 정수 사용
  * 문자열을 정수로 치환하는 함수 필요
  

```python
def recode_sex(sex):
  if sex=='Female':
    return 0
  else:
    return 1
```

####(2) recode_sex 메서드가 반환한 값을 데이터프레임에 sex_color 변수로 추가


```python
tips['sex_color']=tips['sex'].apply(recode_sex)
```

####(3) 산점도에 테이블당 인원수 값 추가
  * 테이블 당 인원수 데이터는 점의 크기로 표현

```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,
    c=tips['sex_color'],
    alpha=0.5)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

###<다양한 종류의 이변량 그래프 그리기>

####(1) seaborn 라이브러리로 산점도 그래프 그리기
  * regplot 메서드 사용
    * 산점도 그래프와 회귀선을 함께 그릴 수 있음
    * fit_reg = False : 회귀선 제거

```python
ax=plt.subplots()
ax=sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax=plt.subplots()
ax=sns.regplot(x='total_bill', y='tip',data=tips, fit_reg=False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

####(2) 산점도 그래프와 히스토그램 한번에 그리기
  * jointplot 메서드

```python
joint=sns.jointplot(x='total_bill',y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill',ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

####(3) 육각 그래프
  *  산점도 그래프는 점이 겹쳐보일 경우 점을구분하기 어려움
  * 2차원 표면에 육각형으로 데이터를 쌓아 표현
  * 특정 데이터의 개수가 많아지면 점점 진한 색으로 표현
  * jointplot 메서드의 kind 인자를 hex로 지정

```python
hexbin=sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10,y=1.03)
```

  * shade=True : 그래프에 음영 효과

```python
ax=plt.subplots()
ax=sns.kdeplot(data=tips['total_bill'],
               data2=tips['tip'],
               shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

####(9) 중복된 그래프 변경
  * 중복된 그래프가 그려지는 위치를 직접 지정하여 원하는 그래프로 교체
  * map_upper : 대각선을 기준으로 위쪽에 그릴 그래프를 지정
  * map_lower : 대각선을 기준으로 아래쪽에 그릴 그래프를 지정
  * map_diag : 대각선을 중심으로 그래프 그리기

```python
pair_grid=sns.PairGrid(tips)
pair_grid=pair_grid.map_upper(sns.regplot)
pari_grid=pair_grid.map_lower(sns.kdeplot)
pari_grid=pair_grid.map_diag(sns.distplot,rug=True)
plt.show()
```

##5. seaborn 라이브러리로 그래프 스타일 설정하기

  * seaborn 라이브러리로 만든 그래프의 스타일은 set_style()을 사용해 변경 가능
  * set_style()을 스타일을 적용할 부분에서 한번만 실행하면 이후의 그래프가 모두 지정한 스타일로 변경됨


###<그래프에 스타일 적용하기>


####(1) 기본 스타일

```python
fig,ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',hue='sex',data=tips, split=True)
```

####(2) whitegrid 스타일
  * 그래프의 배경에 가로 줄

```python
sns.set_style('whitegrid')
fig,ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True)
```

####(3) for문을 사용하여 모든 스타일을 하나씩 적용

```python
fig=plt.figure()
seaborn_styles=['darkgrid','whitegrid','dark','white','ticks']
for idx, style in enumerate(seaborn_styles):
  plot_position=idx+1
  with sns.axes_style(style):
    ax=fig.add_subplot(2,3,plot_position)
    violin=sns.violinplot(x='time',y='total_bill', data=tips, ax=ax)
    violin.set_title(style)

fig.tight_layout()
```
